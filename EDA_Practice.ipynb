```python
# Import liabraries

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Line plot

x =[1,2,3,4,5]
y =[10,45,56,78,56]
#z =[10,20,30,40,50]
plt.plot(x,y)
plt.title("Line plot")
plt.xlabel("data1")
plt.ylabel("data2")
plt.show()
```

```python
# bar plot (works on continuous dataset )

import pandas as pd

data = pd.DataFrame({
    "Class":['A','B','C','D','E'],
    "Students" :[90,46,35,67,55]
})

sns.barplot(x='Class', y='Students', data=data) 
plt.title("Bar Plot") 
plt.xlabel("Class")
plt.ylabel("Students")
plt.show() 
```

Class A having followed Class D and B 

```python
# Count plot

sns.countplot(x='Class', data=data) 
plt.title("Count Plot") 
plt.show()
```

```python
#Histogram 

import numpy as np

data = np.random.randn(100) 
plt.hist(data, bins=10) 
plt.title("Histogram") 
plt.show()
```

```python
# KDE plot

sns.kdeplot(data, shade=True) 
plt.title("KDE Plot") 
plt.show() 
```

```python
#histogram with kde
sns.histplot(data, bins=10, kde=True) 
plt.title("Histogram with KDE") 
plt.show()
```

```python
# Boxplot (To detect outliers)

sns.boxplot(y=data) 
plt.title("Box Plot") 
plt.show() 
```

```python
# Violin plot (Boxplot+KDE)

sns.violinplot(y=data) 
plt.title("Violin Plot") 
plt.show() 
```

```python
# Scatter plot 

x = [1, 2, 3, 4, 5] 
y = [5, 4, 7, 10, 8] 
plt.scatter(x, y) 
plt.title("Scatter Plot") 
plt.show() 
```

```python
# Loading dataset

df = sns.load_dataset('iris') 
df
```

```python
# Pairplot 

sns.pairplot(df, hue='species') 
plt.show() 
```

```python
# Drop Categorical column

df_numeric = df.drop(columns='species')     # ✅ recommended
# or
df_numeric = df.drop('species', axis=1)     # ✅ alternate form
correlation_matrix = df_numeric.corr()
correlation_matrix
```

```python
# Heatmap  

sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm') 
plt.title("Correlation Heatmap")
plt.show()
```

```python
# Pie chart  

labels = ['A', 'B', 'C'] 
sizes = [30, 50, 20] 

plt.pie(sizes, labels=labels, autopct='%1.1f%%',startangle =0)    
plt.title("Pie Chart") 
plt.show() 
```

```python
# Swarm plot

sns.swarmplot(x='species', y='sepal_length', data=df) 
plt.title("Swarm Plot") 
plt.show() 
```

```python
#Strip plot

sns.stripplot(x='species', y='sepal_length', data=df, jitter=True) 
plt.title("Strip Plot") 
plt.show()
```

```python
# Joint plot  (Scatter+Histogram)

sns.jointplot(x='sepal_length', y='petal_length', data=df, kind='scatter') 
plt.show()
```
