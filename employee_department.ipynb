```python
#importing
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#import dataset
ds=pd.read_csv("employee_dataset_for_eda.csv")
ds.head(10)
```

```python
ds.info()
```

```python
ds.describe()
```

```python
#employee count per department
sns.countplot(data=ds, x='Department', palette='Set2')
plt.title('Employee count per Department')
```

```python
#gender distribution
sns.countplot(data=ds, x='Gender', palette='coolwarm')
plt.title('Gender Distribution')
```

```python
#employee location distribution
sns.countplot(data=ds, x='Location', palette='Spectral')
plt.title('Number of Employee by location')
plt.xticks(rotation=45)
```

```python
#age distribution
sns.histplot(ds['Age'], bins=10, kde=True)
plt.title('Age Distribution')
```

```python
#years of Experience distribution
sns.histplot(ds['Years_of_Experience'], bins=15, kde=True)
plt.title('Year of Experience')
```

```python
#monthly salary distribution
sns.boxplot(data=ds, y='Monthly_Salary', color='lightblue')
plt.title('Salary Spread')
```

```python
#average salary by department
sns.barplot(data=ds, x='Department', y='Monthly_Salary', estimator=np.mean, palette='viridis')
plt.title('Average Salary by Department')
```

```python
#performance score by department
sns.boxplot(data=ds, x='Department', y='Performance_Score', palette='pastel')
plt.title('Performance score by Department')
```

```python
#salary distribution by Gender
sns.violinplot(data=ds, x='Gender', y='Monthly_Salary', palette='muted')
plt.title('Salary by Gender')
```

```python
#salary vs experience scatter plot
sns.scatterplot(data=ds, x='Years_of_Experience', y='Monthly_Salary', hue='Department')
plt.title('Salary vs Experience by Department')
```

```python
sns.heatmap(ds[['Age', 'Years_of_Experience', 'Monthly_Salary', 'Performance_Score']].corr(), annot=True,  cmap = 'coolwarm' )
plt.title('Correlation Matrix')
```
